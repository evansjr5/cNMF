## **File: density_downsample.ipynb**
Description: This file performs density based downsampling on anndata <br>
using the pca information. This includes three different approaches: <br>
1. Downsampling all cell types
2. Downsampling only abundant cell types, while keeping rare cell types
3. Downsampling only abundant cell types, and removing rare cell types

### **Imports**

```python
import scanpy as sc
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import anndata as ad
```

### **Import data**

```python
adata = sc.read_h5ad("/home/james/data/immune_exclusion_data/outer_combined_all4_dat.h5ad")
```

## **<span style="color:darkred">Density-dependent downsampling of all cell types</span>**
---
Description: <br>
Down sample all cell types

```python
# Step 1: Compute low-dimensional embedding
sc.tl.pca(adata)

# Use PCA embeddings
embedding = adata.obsm["X_pca"]

# Step 2: Compute local densities using kNN
n_neighbors = 50  # Number of neighbors to consider
nbrs = NearestNeighbors(n_neighbors=n_neighbors).fit(embedding)
distances, indices = nbrs.kneighbors(embedding)

# Inversely of distance to density
density_scores = 1 / (np.mean(distances, axis=1) + 1e-10)

# Normalize density scores to probabilities for sampling
sampling_probabilities = 1 / density_scores  # Higher probabilities for sparse regions
sampling_probabilities /= np.sum(sampling_probabilities)  # Normalize to 1

# Step 3: Perform down-sampling
target_cells = (int(0.2 * adata.n_obs))  # 20% downsampling
selected_indices = np.random.choice(
    np.arange(len(adata)),
    size=target_cells,
    replace=False,
    p=sampling_probabilities,
    )

# Subset data object
adata_down = adata[selected_indices].copy()

#Step  4: Removing stuff
del adata_down.varm["PCs"]
```

```python
# Step 4: Save
# Only use this when needed
#adata_down.write_h5ad("/home/james/data/immune_exclusion_data/pca_density_ds_outer_combined_all4_dat.h5ad")
```

```python
print(adata_down.obs['Cell_Type'].value_counts())
sc.pl.umap(adata_down, color = 'Cell_Type')
```

## **<span style="color:darkred">Density dependent downsampling that protects the rare cell types (<1% of total cells)</span>**
---
Description: <br>
Keep rare cell types at the same counts as in the original data while down sampling the highly observed cell types.


```python
# 1. Identify rare and abundant cell types (1% threshold)
total_cells = adata.n_obs
threshold = total_cells * 0.01
cell_type_counts = adata.obs['Cell_Type'].value_counts()
rare_cell_types = cell_type_counts[cell_type_counts < threshold].index
abundant_cell_types = cell_type_counts[cell_type_counts >= threshold].index

# 2. Separate data
adata_rare = adata[adata.obs['Cell_Type'].isin(rare_cell_types)].copy()
adata_abundant = adata[adata.obs['Cell_Type'].isin(abundant_cell_types)].copy()

# 3. Downsample abundant cells using density-based method

# 3.1 Compute embedding
sc.pp.neighbors(adata_abundant)
sc.tl.pca(adata_abundant)
embedding = adata_abundant.obsm["X_pca"]

# 3.2 kNN and density calculation (same as before, but for adata_abundant)
n_neighbors = 50
nbrs = NearestNeighbors(n_neighbors=n_neighbors).fit(embedding)
distances, indices = nbrs.kneighbors(embedding)
density_scores = 1 / (np.mean(distances, axis=1) + 1e-10)
sampling_probabilities = 1 / density_scores
sampling_probabilities /= np.sum(sampling_probabilities)

# 3.3 Downsampling
target_cells_abundant = int(0.2 * len(adata_abundant)) #20% of abundant cells
selected_indices_abundant = np.random.choice(
    np.arange(len(adata_abundant)),
    size=target_cells_abundant,
    replace=False,
    p=sampling_probabilities,
    )
adata_downsampled_abundant = adata_abundant[selected_indices_abundant].copy()

# 4. Combine rare and downsampled abundant cells
adata_down_rare = ad.concat([adata_downsampled_abundant, adata_rare], join='outer', merge='same')

# Print functions
print(f"Original number of cells: {len(adata)}")
print(f"Downsampled number of cells: {len(adata_downsampled)}")
print(f"Number of rare cell types: {len(rare_cell_types)}")
print(f"Number of abundant cell types: {len(abundant_cell_types)}")
```

```python
# 5. Save
# Only use this when needed
#adata_down_rare.write_h5ad("/home/james/data/immune_exclusion_data/rare_pca_density_ds_outer_combined_all4_dat.h5ad")
```

```python
# 6. Visualize
print(adata_down_rare.obs['Cell_Type'].value_counts())
sc.pl.umap(adata_down_rare, color = 'Cell_Type')
```

## **<span style="color:darkred">Density dependent downsampling that removes rare cell types (<1% total cell count)</span>**
---
Description: <br>
Remove the rare cell types and only perform down sampling on the remaining cells based on a 1% cut off <br>

```python
# 1. Identify rare and abundant cell types (1% threshold)
total_cells = adata.n_obs
threshold = total_cells * 0.01
cell_type_counts = adata.obs['Cell_Type'].value_counts()  # Assuming 'Cell_Type' is your cell type column
rare_cell_types = cell_type_counts[cell_type_counts < threshold].index
abundant_cell_types = cell_type_counts[cell_type_counts >= threshold].index

# 2. Separate data
adata_rare = adata[adata.obs['Cell_Type'].isin(rare_cell_types)].copy()
adata_abundant = adata[adata.obs['Cell_Type'].isin(abundant_cell_types)].copy()

# 3. Downsample abundant cells using density-based method

# Compute embedding
sc.pp.neighbors(adata_abundant)
sc.tl.pca(adata_abundant)
embedding = adata_abundant.obsm["X_pca"]

# kNN and density calculation
n_neighbors = 50
nbrs = NearestNeighbors(n_neighbors=n_neighbors).fit(embedding)
distances, indices = nbrs.kneighbors(embedding)
density_scores = 1 / (np.mean(distances, axis=1) + 1e-10)
sampling_probabilities = 1 / density_scores
sampling_probabilities /= np.sum(sampling_probabilities)

# Downsampling (adjust target_cells as needed)
target_cells_abundant = int(0.2 * len(adata_abundant)) #20% of abundant cells
selected_indices_abundant = np.random.choice(
    np.arange(len(adata_abundant)),
    size=target_cells_abundant,
    replace=False,
    p=sampling_probabilities,
)
adata_down_filter = adata_abundant[selected_indices_abundant].copy()

print(f"Original number of cells: {len(adata)}")
print(f"Downsampled number of cells: {len(adata_downsampled)}")
print(f"Number of rare cell types: {len(rare_cell_types)}")
print(f"Number of abundant cell types: {len(abundant_cell_types)}")
```

```python
# 6. Visualize
print(adata_down_filter.obs['Cell_Type'].value_counts())
sc.pl.umap(adata_down_filter, color = 'Cell_Type')
```
